# Features generation

```python
from scipy.stats import kurtosis,skew,entropy
```

```python
r_mean = []
g_mean = []
b_mean = []
for i in range(train_x.shape[0]):
    img = train_x[i]
    r_mean.append(np.mean(img[:,:,0]))
    g_mean.append(np.mean(img[:,:,1]))
    b_mean.append(np.mean(img[:,:,2]))
```

```python
r_std = []
g_std = []
b_std = []
for i in range(train_x.shape[0]):
    img = train_x[i]
    r_std.append(np.std(img[:,:,0]))
    g_std.append(np.std(img[:,:,1]))
    b_std.append(np.std(img[:,:,2]))
```

```python
r_kur = []
g_kur = []
b_kur = []
for i in range(train_x.shape[0]):
    img = train_x[i]
    r_kur.append(np.mean(kurtosis(img, axis = 0, fisher=True, bias=True)))
    g_kur.append(np.mean(kurtosis(img, axis = 1, fisher=True, bias=True)))
    b_kur.append(np.mean(kurtosis(img, axis = 2, fisher=True, bias=True)))
```

```python
r_skew = []
g_skew = []
b_skew = []
for i in range(train_x.shape[0]):
    img = train_x[i]
    r_skew.append(np.mean(skew(img, axis = 0,  bias=True)))
    g_skew.append(np.mean(skew(img, axis = 1,  bias=True)))
    b_skew.append(np.mean(skew(img, axis = 2,  bias=True)))
```

```python
r_entropy = []
g_entropy = []
b_entropy = []
for i in range(train_x.shape[0]):
    img = train_x[i]
    r_entropy.append(np.mean(entropy(img, axis = 0 )))
    g_entropy.append(np.mean(entropy(img, axis = 1 )))
    b_entropy.append(np.mean(entropy(img, axis = 2)))
```

```python
r_range = []
g_range = []
b_range = []
for i in range(train_x.shape[0]):
    img = train_x[i]
    r_range.append(np.max(img[:,:,0]) - np.min(img[:,:,0]))
    g_range.append(np.max(img[:,:,1]) - np.min(img[:,:,1]))
    b_range.append(np.max(img[:,:,2]) - np.min(img[:,:,2]))
```

```python
lists = []
lists.append(r_mean)
lists.append(b_mean)
lists.append(g_mean)

lists.append(r_std)
lists.append(b_std)
lists.append(g_std)

lists.append(r_kur)
lists.append(b_kur)
lists.append(g_kur)

lists.append(r_skew)
lists.append(b_skew)
lists.append(g_skew)

lists.append(r_entropy)
lists.append(b_entropy)
lists.append(g_entropy)

lists.append(r_range)
lists.append(b_range)
lists.append(g_range)
```
